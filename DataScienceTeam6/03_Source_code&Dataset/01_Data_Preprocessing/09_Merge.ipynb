```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Read all dataes

Gas_Station_Price_History_drop = pd.read_excel("../00_Data/Transformed_data/Gas_Station_Price_History_drop.xlsx")
GDP_data_cleaned = pd.read_excel("../00_Data/Transformed_data/GDP_data_cleaned.xlsx")
Interpolate_Import_and_Export_drop = pd.read_excel("../00_Data/Transformed_data/Interpolate_Import_and_Export_drop.xlsx")
interpolation_function_Consumer_Price_index_drop = pd.read_excel("../00_Data/Transformed_data/interpolation_function_Consumer_Price_index_drop.xlsx")
interpolation_function_Exchange_Rate_drop = pd.read_excel("../00_Data/Transformed_data/interpolation_function_Exchange_Rate_drop.xlsx")
new_copy_Base_Interest_Rate_drop = pd.read_excel("../00_Data/Transformed_data/new_copy_Base_Interest_Rate_drop.xlsx")
traffic_rate_drop_calculated = pd.read_excel("../00_Data/Transformed_data/traffic_rate_drop_calculated.xlsx")
interpolation_function_National_Car_Ownership = pd.read_excel('../00_Data/Transformed_data/interpolation_function_National_Car_Ownership.xlsx')
```

```python
#Integrated each data.

df=traffic_rate_drop_calculated.copy()

df['Base Interest Rate'] = new_copy_Base_Interest_Rate_drop['Base Interest Rate']
df['Exchange Rate'] = interpolation_function_Exchange_Rate_drop['Exchange Rate']
df['Gross Domestic Product (GDP)'] = GDP_data_cleaned['Gross Domestic Product (GDP)']
df['Economic Growth Rate (Real GDP Growth Rate)'] = GDP_data_cleaned['Economic Growth Rate (Real GDP Growth Rate)']
df['Export price'] = Interpolate_Import_and_Export_drop['Export price']
df['Export growth rate'] = Interpolate_Import_and_Export_drop['Export growth rate']
df['Import price'] = Interpolate_Import_and_Export_drop['Import price']
df['Import growth rate'] = Interpolate_Import_and_Export_drop['Import growth rate']
df['Nationwide Consumer Price Index'] = interpolation_function_Consumer_Price_index_drop['Nationwide Consumer Price Index']
df['Exchange Rate'] = interpolation_function_Exchange_Rate_drop['Exchange Rate']
df['Regular Unleaded Gasoline'] = Gas_Station_Price_History_drop['Regular Unleaded Gasoline']
df['Diesel Fuel for Automobiles'] = Gas_Station_Price_History_drop['Diesel Fuel for Automobiles']
df['number of car ownership'] = interpolation_function_National_Car_Ownership['number of car ownership']
```

```python
df_cleaned = df.select_dtypes(include=['number'])

# Calculating the correlation matrix
corr_matrix = df_cleaned.corr()

# Drawing the heatmap
plt.figure(figsize=(12, 8))
heatmap = sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', linewidths=0.5, fmt='.2f')
plt.title('Correlation Heatmap')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x='Total Traffic Volume', data=df)
plt.title('Boxplot of Total Traffic Volume')
plt.show()

# Outlier Detection
q1 = df['Total Traffic Volume'].quantile(0.25)
q3 = df['Total Traffic Volume'].quantile(0.75)
iqr = q3 - q1
lower_bound = q1 - 1.5 * iqr
upper_bound = q3 + 1.5 * iqr

outliers = df[(df['Total Traffic Volume'] < lower_bound) | (df['Total Traffic Volume'] > upper_bound)]
print("Detected outliers:\n", len(outliers))
print(outliers)

# Outlier Adjustment (Example: Replacing outliers with upper or lower bound)
df['Total Traffic Volume'].loc[df['Total Traffic Volume'] > upper_bound] = upper_bound
df['Total Traffic Volume'].loc[df['Total Traffic Volume'] < lower_bound] = lower_bound

# Visualizing adjusted data
plt.figure(figsize=(10, 6))
sns.boxplot(x='Total Traffic Volume', data=df)
plt.title('Boxplot of Total Traffic Volume after outlier handling')
plt.show()
```
