```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
traffic_rate = pd.read_excel("../00_Data/Original_data/traffic_rate.xlsx")
traffic_rate.head()
```

```python
traffic_rate.columns
```

```python
traffic_rate.info()
```

```python
traffic_rate_drop = traffic_rate.dropna()
traffic_rate_drop
```

```python
traffic_rate_drop.info()
```

```python
traffic_rate_drop.describe()
```

```python
traffic_rate_drop.dtypes
```

```python
traffic_rate_drop_calculated = traffic_rate_drop.groupby('Date')['Total Traffic Volume'].sum().reset_index()
traffic_rate_drop_calculated
```

```python
plt.figure(figsize=(10, 5))
plt.plot(traffic_rate_drop_calculated['Date'], traffic_rate_drop_calculated['Total Traffic Volume'], color='blue')
plt.title('Total Traffic Volume Over Time')
plt.xlabel('Date')
plt.ylabel('Total Traffic Volume')
```

```python
traffic_rate_drop_calculated['Date'] = pd.to_datetime(traffic_rate_drop_calculated['Date'], format='%Y%m%d')
traffic_rate_drop_calculated['Date'] = traffic_rate_drop_calculated['Date'].dt.strftime('%Y-%m-%d')
traffic_rate_drop_calculated.to_excel('../00_Data/Transformed_data/traffic_rate_drop_calculated.xlsx', index=False)
```
